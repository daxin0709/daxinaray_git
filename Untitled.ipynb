```python
import otdrparser

def sor_data(file_path):
    with open(file_path, 'rb') as f:
        sor = otdrparser.SOR(f.read())
    
    # 提取距離 (Distance) 與 功率 (Loss/Power)
    # 注意：通常需要根據檔案中的偏移量與縮放比例轉換
    data_points = sor.data_points
    events = sor.events
    
    return data_points, events
```

```python
import matplotlib.pyplot as plt

def plot_otdr(data_points, events):
    distances = [p.distance for p in data_points]
    powers = [p.power for p in data_points]

    plt.figure(figsize=(12, 6))
    plt.plot(distances, powers, label='OTDR Trace', color='blue', lw=1)
    
    # 標註事件點 (例如接續損耗或反射)
    for event in events:
        plt.axvline(x=event.distance, color='red', linestyle='--', alpha=0.5)
        plt.text(event.distance, max(powers), f"E: {event.distance}m", rotation=90)

    plt.title("Detailed OTDR Trace Analysis")
    plt.xlabel("Distance (m)")
    plt.ylabel("Reflectance (dB)")
    plt.grid(True)
    plt.show()
```

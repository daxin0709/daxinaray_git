```python
import otdrparser

def sor_file(file_path):
    with open(file_path, 'rb') as f:
        # 解析二進位內容
        sor = otdrparser.SOR(f.read())
    
    print("--- .sor 檔案讀取成功 ---")
    # 1. 基本資訊
    print(f"測試波長: {sor.fixed_parameters.wavelength} nm")
    print(f"脈衝寬度: {sor.fixed_parameters.pulse_width} ns")
    print(f"解析度: {sor.data_points.spacing} meters")
    print(f"總數據點數: {len(sor.data_points.values)}")
    
    # 2. 列印事件表 (Events)
    print(f"\n--- 偵測到 {len(sor.key_events.events)} 個事件 ---")
    for i, event in enumerate(sor.key_events.events):
        print(f"事件 {i+1}:")
        print(f"  距離: {event.distance:.2f} m")
        print(f"  損耗: {event.loss:.3f} dB")
        print(f"  反射: {event.reflectance:.2f} dB")
        print(f"  類型: {event.event_type}")

    return sor

# 使用方法
# sor_data = analyze_sor_file("your_file.sor")
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_sor_trace(sor):
    # 提取數據點
    raw_values = sor.data_points.values
    spacing = sor.data_points.spacing / 1000  # 轉換為公里 (km)
    
    # 生成距離軸: [0, spacing, 2*spacing, ...]
    distances = np.arange(len(raw_values)) * spacing
    
    # 建立畫布
    plt.figure(figsize=(15, 7))
    
    # 繪製主曲線
    plt.plot(distances, raw_values, label='OTDR Trace', color='#2c3e50', linewidth=1)
    
    # 標註事件點
    for event in sor.key_events.events:
        dist_km = event.distance / 1000
        # 在圖上畫紅色垂直線標記事件
        plt.axvline(x=dist_km, color='red', linestyle='--', alpha=0.4)
        # 在線上方寫文字說明
        plt.text(dist_km, max(raw_values), f" {event.event_type}\n {dist_km:.3f}km", 
                 fontsize=8, verticalalignment='top')

    # 優化圖表外觀
    plt.title("OTDR Optical Fiber Analysis", fontsize=14)
    plt.xlabel("Distance (km)", fontsize=12)
    plt.ylabel("Reflectance / Power (dB)", fontsize=12)
    plt.grid(True, which='both', linestyle=':', alpha=0.6)
    plt.legend()
    
    # 顯示圖形
    plt.show()

# 呼叫函數
# plot_sor_trace(sor_data)
```

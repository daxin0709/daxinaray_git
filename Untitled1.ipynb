```python
import otdrparser
import json
import matplotlib.pyplot as plt
import numpy as np

def process_otdr_data(sor_file_path):
    # 1. 讀取並解析 .sor 檔案
    try:
        with open(sor_file_path, 'rb') as f:
            # otdrparser 會自動識別 Telcordia 格式
            sor_data = otdrparser.sor_to_dict(f.read())
        
        # 2. 提取關鍵數據：Data Points (點位數據)
        # 通常在 'data_points' 鍵值下，包含 'distance' 和 'reflectance'
        points = sor_data.get('data_points', [])
        distances = [p['distance'] for p in points]
        reflectance = [p['reflectance'] for p in points]

        # 3. 儲存為 JSON (方便之後 AI 模型讀取)
        json_filename = sor_file_path.replace('.sor', '.json')
        with open(json_filename, 'w', encoding='utf-8') as jf:
            json.dump(sor_data, jf, indent=4, ensure_ascii=False)
        
        print(f"✅ 解析成功！JSON 已儲存至: {json_filename}")
        
        # 4. 視覺化：畫出 OTDR 曲線圖
        plt.figure(figsize=(12, 6))
        plt.plot(distances, reflectance, label='OTDR Trace', color='blue', linewidth=1)
        plt.title(f"OTDR Trace Analysis: {sor_file_path}")
        plt.xlabel("Distance (km)")
        plt.ylabel("Reflectance / Loss (dB)")
        plt.grid(True, linestyle='--', alpha=0.7)
        plt.legend()
        plt.show()
        
        return sor_data

    except Exception as e:
        print(f"❌ 發生錯誤: {e}")
        return None

# --- 執行測試 ---
# 請確保你的 data_raw 資料夾裡有放一個測試檔，並修改檔名
# my_data = process_otdr_data('data_raw/sample_test.sor')
```

```python
import otdrparser
import pandas as pd
import matplotlib.pyplot as plt
import os

# --- 設定檔案路徑 ---
target_file = 'data_raw/data_rawsample_test.sor'

def process_otdr():
    print(f"1. 開始處理檔案：{target_file}")
    
    if not os.path.exists(target_file):
        print(f"❌ 錯誤：找不到檔案 {target_file}")
        return

    try:
        # 讀取檔案
        with open(target_file, 'rb') as f:
            # 使用 otdrparser 解析檔案物件
            sor_data = otdrparser.parse(f)
        
        # 2. 尋找 DataPts 數據區塊
        data_pts_block = None
        for block in sor_data:
            if block.get('name') == 'DataPts':
                data_pts_block = block
                break
        
        if data_pts_block:
            print("✅ 成功找到 DataPts 數據區塊！")
            
            # 3. 提取原始點位
            points = data_pts_block.get('data_points', [])
            df = pd.DataFrame(points)
            
            if not df.empty:
                # --- 自動偵測欄位名稱 ---
                print(f"目前所有的欄位有: {list(df.columns)}")
                
                # 找出 X 軸 (通常是 distance, dist 或 index)
                x_col = 'distance' if 'distance' in df.columns else df.columns[0]
                
                # 找出 Y 軸 (通常是 reflectance, loss, 或 data)
                # 我們找除了 X 軸以外的第二個欄位
                y_col = 'reflectance' if 'reflectance' in df.columns else (df.columns[1] if len(df.columns) > 1 else df.columns[0])
                
                print(f"✅ 偵測到 X軸: {x_col}, Y軸: {y_col}")

               # --- 4. 數據校正 (還原儀器視角) ---
                # A. 先將原始數據轉換為 dB 浮點數
                df[y_col] = df[y_col] / 1.0
                
                # B. 數值反轉 (將負的功率趨勢轉為正的損耗趨勢)
                # 這裡我們使用一個補償係數 (Offset)
                # 根據你的圖 2，起始點約在 14dB，而目前圖 1 起始點在 -51dB 左右
                # 所以補償值約為 14 - (-51) = 65
                
                offset_value = 65.0 
                df[y_col] = df[y_col] + offset_value

                print(f"✅ 數值校正完成。起始點現在約為: {df[y_col].iloc[0]:.2f} dB")
                
                # --- 5. 視覺化繪圖 ---
                plt.figure(figsize=(12, 6))
                plt.plot(df[x_col], df[y_col], color='blue', label='Corrected Trace', linewidth=1)
                
                plt.title(f"OTDR Trace: {os.path.basename(target_file)}")
                plt.xlabel(f"{x_col} (km)")
                plt.ylabel(f"{y_col} (dB)")
                plt.grid(True, linestyle='--', alpha=0.6)
                plt.legend()
                plt.show()
                # --- 6. 儲存為 JSON (供 AI 模型使用) ---
                json_path = target_file.replace('.sor', '.json')
                df.to_json(json_path, orient='records')
                print(f"✅ 已儲存校正後的數據至: {json_path}")
                
            else:
                print("❌ DataPts 區塊內沒有點位數據。")
        else:
            print("❌ 在檔案中找不到 DataPts 區塊。")
            print("現有的區塊名稱有:", [b.get('name') for b in sor_data])

    except Exception as e:
        print(f"❌ 執行過程發生錯誤: {e}")

# 執行主程式
process_otdr()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

def extract_fiber_features(df):
    """
    自動偵測欄位、進行公里換算並提取特徵
    """
    features = {}
    
    if df is None or df.empty:
        return None

    # 1. 自動判定 X 和 Y 欄位
    x_col = df.columns[0]
    y_col = df.columns[-1]
    
    # --- 關鍵修正：將距離從 公尺(m) 轉換為 公里(km) ---
    # 如果最大值大於 500，通常代表單位是公尺而非公里
    if df[x_col].max() > 500:
        df[x_col] = df[x_col] / 1000.0
        print(f"🔄 偵測到原始單位為公尺，已自動轉換為公里 (km)")

    # 2. 計算總長度 (km)
    features['total_length'] = df[x_col].max()
    
    # 3. 計算總損耗 (dB)
    features['total_loss'] = abs(df[y_col].iloc[-1] - df[y_col].iloc[0])
    
    # 4. 計算平均斜率 (dB/km)
    # 取 20% 到 80% 區間以避開死區與末端
    lower_bound = features['total_length'] * 0.2
    upper_bound = features['total_length'] * 0.8
    
    mid_data = df[(df[x_col] > lower_bound) & (df[x_col] < upper_bound)]
    
    if not mid_data.empty:
        model = LinearRegression()
        X = mid_data[[x_col]].values
        y = mid_data[y_col].values
        model.fit(X, y)
        features['avg_slope'] = abs(model.coef_[0])
    else:
        features['avg_slope'] = 0.0
        
    return features

# --- 執行測試 ---
try:
    fiber_info = extract_fiber_features(df)
    
    if fiber_info:
        print("\n📊 修正後的正確特徵：")
        print(f"📏 總長度: {fiber_info['total_length']:.3f} km")
        print(f"📉 總損耗: {fiber_info['total_loss']:.3f} dB")
        print(f"📉 平均衰減率: {fiber_info['avg_slope']:.4f} dB/km")
        
        # 根據 G.652 標準，1550nm 正常衰減約 0.18 ~ 0.25 dB/km
        if fiber_info['avg_slope'] > 0.25:
            print("⚠️ 判定結果: 衰減異常")
        else:
            print("✅ 判定結果: 衰減正常")

except Exception as e:
    print(f"❌ 執行出錯: {e}")
```

```python
import os
import pandas as pd
import otdrparser

# 1. 設定路徑與參數
source_folder = 'data_raw/'
results = []
offset_value = 65.0  # 這是我們之前調校出來的完美數值

print(f"🚀 開始批量處理資料夾: {source_folder}")

# 2. 取得所有 .sor 檔案
sor_files = [f for f in os.listdir(source_folder) if f.endswith('.sor')]

for filename in sor_files:
    file_path = os.path.join(source_folder, filename)
    try:
        # --- A. 讀取與解析 ---
        with open(file_path, 'rb') as f:
            sor_data = otdrparser.parse(f)
        
        # 尋找 DataPts
        data_pts_block = next((b for b in sor_data if b.get('name') == 'DataPts'), None)
        if not data_pts_block: continue
            
        df_temp = pd.DataFrame(data_pts_block.get('data_points', []))
        
        # --- B. 數據校正 (Y軸與X軸) ---
        x_col, y_col = df_temp.columns[0], df_temp.columns[-1]
        
        # 距離轉公里
        if df_temp[x_col].max() > 500:
            df_temp[x_col] = df_temp[x_col] / 1000.0
            
        # 功率轉損耗並對齊儀器
        df_temp[y_col] = (df_temp[y_col] / 1000.0) + offset_value
        
        # --- C. 提取特徵 ---
        # 這裡直接呼叫你剛才寫好的功能 (假設已定義 extract_fiber_features)
        info = extract_fiber_features(df_temp)
        
        if info:
            info['file_name'] = filename
            # 這裡可以根據檔名自動打標籤 (如果檔名包含 Normal, Bend 等)
            if 'normal' in filename.lower(): info['label'] = 'Normal'
            elif 'bend' in filename.lower(): info['label'] = 'Bend'
            elif 'broken' in filename.lower(): info['label'] = 'Broken'
            else: info['label'] = 'Unknown'
            
            results.append(info)
            print(f"✅ 已處理: {filename} ({info['label']})")

    except Exception as e:
        print(f"❌ 檔案 {filename} 處理失敗: {e}")

# 3. 儲存結果
final_df = pd.DataFrame(results)
final_df.to_csv('fiber_dataset.csv', index=False)
print(f"\n✨ 全部處理完成！已產出 fiber_dataset.csv，共 {len(final_df)} 筆數據。")
```

```python
def auto_label_logic(info, expected_length=14.5):
    """
    根據物理特徵自動打標籤的邏輯
    expected_length: 這條光纖理論上的標準長度(km)
    """
    # 1. 判定 Broken (長度顯著縮短)
    # 如果偵測長度比預期短了 5% 以上，初步判定為斷線
    if info['total_length'] < (expected_length * 0.95):
        return 'Broken'
    
    # 2. 判定 Bend (長度對了，但損耗或斜率太高)
    # 正常 G.652 衰減在 0.25 以下
    if info['avg_slope'] > 0.28:
        return 'Bend'
    
    # 3. 判定 Normal
    return 'Normal'

# 在批量處理迴圈中使用：
# info['label'] = auto_label_logic(info, expected_length=14.9)
```

```python
import os
import pandas as pd
import otdrparser
from sklearn.linear_model import LinearRegression

# --- 功能 1：定義特徵提取與智能判定函數 ---
def extract_and_classify(df, filename):
    features = {}
    # 假設第 0 欄是距離，最後一欄是損耗功率
    x_col = df.columns[0]
    y_col = df.columns[-1]

    # A. 單位換算：公里 (km)
    x_raw = df[x_col].values
    if x_raw.max() > 500: 
        x_km = x_raw / 1000.0
    else:
        x_km = x_raw
    
    # B. 取得 Y 軸數據 (dB)
    y_db = df[y_col].values
    total_len = x_km.max()
    
    # C. 計算平均衰減率 (dB/km) - 避開反射峰與雜訊區
    mask = (x_km > total_len * 0.2) & (x_km < total_len * 0.8)
    x_mid = x_km[mask].reshape(-1, 1)
    y_mid = y_db[mask]

    avg_slope = 0.0
    if len(x_mid) > 1:
        model = LinearRegression()
        model.fit(x_mid, y_mid)
        # 取絕對值，因為衰減是負斜率
        avg_slope = abs(model.coef_[0])

    # D. 核心判定邏輯 (基於你的數據分佈微調)
    if avg_slope > 0.4:
        label = 'Bend'         
    elif avg_slope < 0.12:
        # 如果長度明顯縮短 (假設目標長度是 15km 或 25km)
        if total_len < 10.0:   
            label = 'Broken'   
        else:
            label = 'Abnormal_Low' 
    else:
        label = 'Normal'       

    # 封裝結果
    features['file_name'] = filename
    features['total_length'] = round(total_len, 3)
    features['avg_slope'] = round(avg_slope, 4)
    features['label'] = label
    
    return features

# --- 功能 2：批量執行生產線 ---
source_folder = 'data_raw/'
offset_value = 65.0  # 校正補償值
all_results = []

# 檢查資料夾是否存在
if not os.path.exists(source_folder):
    print(f"❌ 找不到資料夾: {source_folder}，請確認資料夾名稱正確。")
else:
    # 掃描資料夾內所有 .sor 檔案
    sor_files = [f for f in os.listdir(source_folder) if f.endswith('.sor')]
    print(f"🚀 開始處理 {len(sor_files)} 個檔案...")

    for filename in sor_files:
        try:
            file_path = os.path.join(source_folder, filename)
            with open(file_path, 'rb') as f:
                sor_data = otdrparser.parse(f)
            
            # 尋找數據塊
            data_pts = next((b for b in sor_data if b.get('name') == 'DataPts'), None)
            if data_pts:
                df_temp = pd.DataFrame(data_pts.get('data_points', []))
                
                # 數據校正 (mdB 轉 dB 並加上 Offset)
                y_col = df_temp.columns[-1]
                df_temp[y_col] = (df_temp[y_col] / 1.0) + offset_value
                
                # 呼叫特徵提取函數
                file_info = extract_and_classify(df_temp, filename)
                all_results.append(file_info)
                
        except Exception as e:
            print(f"⚠️ 檔案 {filename} 處理出錯: {e}")

    # --- 功能 3：產出數據集 ---
    if all_results:
        final_df = pd.DataFrame(all_results)
        # 存檔前請確認 Excel 已關閉
        try:
            final_df.to_csv('fiber_dataset.csv', index=False)
            print(f"✨ 處理完成！已產出 fiber_dataset.csv，共有 {len(final_df)} 筆特徵數據。")
            print("\n📊 各類別統計結果：")
            print(final_df['label'].value_counts())
        except PermissionError:
            print("❌ 存檔失敗！請先關閉正在開啟中的 fiber_dataset.csv 檔案。")
    else:
        print("❌ 搜尋不到有效的 DataPts 數據。")
```

```python
import os
import pandas as pd
import otdrparser
from sklearn.linear_model import LinearRegression

# --- 功能 1：升級版特徵提取與事件診斷函數 ---
def extract_and_classify_v3(sor_data, df, filename):
    features = {}
    x_col = df.columns[0]
    y_col = df.columns[-1]

    # A. 距離與損耗校正
    x_raw = df[x_col].values
    x_km = x_raw / 1000.0 if x_raw.max() > 500 else x_raw
    y_db = df[y_col].values
    total_len = x_km.max()
    
    # B. 計算平均衰減率 (dB/km)
    mask = (x_km > total_len * 0.2) & (x_km < total_len * 0.8)
    avg_slope = 0.0
    if len(df[mask]) > 1:
        model = LinearRegression()
        model.fit(x_km[mask].reshape(-1, 1), y_db[mask])
        avg_slope = abs(model.coef_[0])

    # C. 【新增】解析事件點資訊 (定位障礙)
    event_list = next((b for b in sor_data if b.get('name') == 'Event'), None)
    fault_location = None
    max_event_loss = 0.0
    event_count = 0
    
    if event_list and 'events' in event_list:
        events = event_list['events']
        event_count = len(events)
        for ev in events:
            # 取得事件位置與損耗 (mdB 轉 dB)
            pos = ev.get('distance', 0) / 1000.0 if ev.get('distance', 0) > 500 else ev.get('distance', 0)
            loss = ev.get('loss', 0) / 1000.0
            
            # 避開末端反射，抓取中間損耗最大的點 (判定門檻設為 0.3 dB)
            if loss > 0.3 and pos < total_len * 0.98:
                if loss > max_event_loss:
                    max_event_loss = loss
                    fault_location = round(pos, 3)

    # D. 智能判定邏輯 (結合事件診斷)
    if avg_slope > 0.4:
        label = 'Bend'
    elif total_len < 10.0: # 假設標準長度遠大於 10km
        label = 'Broken'
    elif fault_location is not None and max_event_loss > 0.5:
        label = 'Point_Fault'  # 判定為特定點故障
    else:
        label = 'Normal'

    # 封裝所有輸出資訊
    features['file_name'] = filename
    features['total_length'] = round(total_len, 3)
    features['avg_slope'] = round(avg_slope, 4)
    features['label'] = label
    features['event_count'] = event_count
    features['fault_at_km'] = fault_location      # 障礙公里數
    features['fault_loss_db'] = round(max_event_loss, 3) # 障礙點損耗值
    
    return features

# --- 功能 2：執行批量生產線 ---
source_folder = 'data_raw/'
offset_value = 65.0  
all_results = []

if not os.path.exists(source_folder):
    print(f"❌ 找不到資料夾: {source_folder}")
else:
    sor_files = [f for f in os.listdir(source_folder) if f.endswith('.sor')]
    print(f"🚀 啟動診斷生產線，目標檔案數： {len(sor_files)}")

    for filename in sor_files:
        try:
            file_path = os.path.join(source_folder, filename)
            with open(file_path, 'rb') as f:
                sor_data = otdrparser.parse(f)
            
            data_pts = next((b for b in sor_data if b.get('name') == 'DataPts'), None)
            if data_pts:
                df_temp = pd.DataFrame(data_pts.get('data_points', []))
                y_col = df_temp.columns[-1]
                # mdB 轉 dB 校正
                df_temp[y_col] = (df_temp[y_col] / 1.0) + offset_value
                
                # 傳入原始 sor_data 以便解析事件區塊
                file_info = extract_and_classify_v3(sor_data, df_temp, filename)
                all_results.append(file_info)
        except Exception as e:
            print(f"⚠️ 檔案 {filename} 處理出錯: {e}")

    # --- 功能 3：匯出結果 ---
    if all_results:
        final_df = pd.DataFrame(all_results)
        try:
            final_df.to_csv('fiber_dataset_diagnostic.csv', index=False)
            print(f"✨ 診斷完成！已產出 fiber_dataset_diagnostic.csv")
            # 顯示簡單報表
            print("\n--- 診斷摘要 ---")
            print(final_df['label'].value_counts())
            # 顯示前三個發現障礙的檔案
            faults = final_df[final_df['label'] == 'Point_Fault']
            if not faults.empty:
                print("\n🚩 障礙點偵測範例：")
                print(faults[['file_name', 'fault_at_km', 'fault_loss_db']].head(3))
        except PermissionError:
            print("❌ 錯誤：請先關閉 Excel 中的 CSV 檔案！")
```

```python
def extract_and_classify_v4(df, filename):
    features = {}
    x_col = df.columns[0]
    y_col = df.columns[-1]

    # 1. 單位與基本數據準備
    x_km = df[x_col].values / 1000.0 if df[x_col].max() > 500 else df[x_col].values
    y_db = df[y_col].values
    total_len = x_km.max()
    
    # 2. 計算平均衰減 (Slope)
    mask = (x_km > total_len * 0.2) & (x_km < total_len * 0.8)
    avg_slope = 0.0
    if len(df[mask]) > 1:
        model = LinearRegression()
        model.fit(x_km[mask].reshape(-1, 1), y_db[mask])
        avg_slope = abs(model.coef_[0])

    # 3. 【自研演算法】手動偵測突發事件 (Event Detection)
    # 計算相鄰點之間的損耗差
    y_diff = -pd.Series(y_db).diff() # 損耗通常是負值，取負變正
    
    # 找尋損耗大於 0.5 dB 的點 (門檻可調)
    significant_events = y_diff[y_diff > 0.5]
    
    fault_location = None
    max_event_loss = 0.0
    
    if not significant_events.empty:
        # 排除前 5% 和 後 5% (避免抓到死區或末端反射)
        valid_indices = significant_events.index[
            (significant_events.index > len(df) * 0.05) & 
            (significant_events.index < len(df) * 0.95)
        ]
        if not valid_indices.empty:
            idx = significant_events[valid_indices].idxmax()
            max_event_loss = significant_events[idx]
            fault_location = round(x_km[idx], 3)

    # 4. 判定邏輯
    if total_len < 10.0:  # 比預期長度短很多
        label = 'Broken'
    elif avg_slope > 0.6: # 整條線衰減都很高
        label = 'Severe_Bend'
    elif fault_location is not None:
        label = 'Point_Fault' # 只有特定某個點掉很多
    else:
        label = 'Normal'

    features['file_name'] = filename
    features['total_length'] = round(total_len, 3)
    features['avg_slope'] = round(avg_slope, 4)
    features['label'] = label
    features['fault_at_km'] = fault_location
    features['fault_loss_db'] = round(max_event_loss, 3)
    
    return features
```

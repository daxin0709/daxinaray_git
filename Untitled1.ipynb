```python
import otdrparser
import json
import matplotlib.pyplot as plt
import numpy as np

def process_otdr_data(sor_file_path):
    # 1. 讀取並解析 .sor 檔案
    try:
        with open(sor_file_path, 'rb') as f:
            # otdrparser 會自動識別 Telcordia 格式
            sor_data = otdrparser.sor_to_dict(f.read())
        
        # 2. 提取關鍵數據：Data Points (點位數據)
        # 通常在 'data_points' 鍵值下，包含 'distance' 和 'reflectance'
        points = sor_data.get('data_points', [])
        distances = [p['distance'] for p in points]
        reflectance = [p['reflectance'] for p in points]

        # 3. 儲存為 JSON (方便之後 AI 模型讀取)
        json_filename = sor_file_path.replace('.sor', '.json')
        with open(json_filename, 'w', encoding='utf-8') as jf:
            json.dump(sor_data, jf, indent=4, ensure_ascii=False)
        
        print(f"✅ 解析成功！JSON 已儲存至: {json_filename}")
        
        # 4. 視覺化：畫出 OTDR 曲線圖
        plt.figure(figsize=(12, 6))
        plt.plot(distances, reflectance, label='OTDR Trace', color='blue', linewidth=1)
        plt.title(f"OTDR Trace Analysis: {sor_file_path}")
        plt.xlabel("Distance (km)")
        plt.ylabel("Reflectance / Loss (dB)")
        plt.grid(True, linestyle='--', alpha=0.7)
        plt.legend()
        plt.show()
        
        return sor_data

    except Exception as e:
        print(f"❌ 發生錯誤: {e}")
        return None

# --- 執行測試 ---
# 請確保你的 data_raw 資料夾裡有放一個測試檔，並修改檔名
# my_data = process_otdr_data('data_raw/sample_test.sor')
```

```python
import otdrparser
import pandas as pd
import matplotlib.pyplot as plt
import os

# --- 設定檔案路徑 ---
target_file = 'data_raw/data_rawsample_test.sor'

def process_otdr():
    print(f"1. 開始處理檔案：{target_file}")
    
    if not os.path.exists(target_file):
        print(f"❌ 錯誤：找不到檔案 {target_file}")
        return

    try:
        # 讀取檔案
        with open(target_file, 'rb') as f:
            # 使用 otdrparser 解析檔案物件
            sor_data = otdrparser.parse(f)
        
        # 2. 尋找 DataPts 數據區塊
        data_pts_block = None
        for block in sor_data:
            if block.get('name') == 'DataPts':
                data_pts_block = block
                break
        
        if data_pts_block:
            print("✅ 成功找到 DataPts 數據區塊！")
            
            # 3. 提取原始點位
            points = data_pts_block.get('data_points', [])
            df = pd.DataFrame(points)
            
            if not df.empty:
                # --- 自動偵測欄位名稱 ---
                print(f"目前所有的欄位有: {list(df.columns)}")
                
                # 找出 X 軸 (通常是 distance, dist 或 index)
                x_col = 'distance' if 'distance' in df.columns else df.columns[0]
                
                # 找出 Y 軸 (通常是 reflectance, loss, 或 data)
                # 我們找除了 X 軸以外的第二個欄位
                y_col = 'reflectance' if 'reflectance' in df.columns else (df.columns[1] if len(df.columns) > 1 else df.columns[0])
                
                print(f"✅ 偵測到 X軸: {x_col}, Y軸: {y_col}")

               # --- 4. 數據校正 (還原儀器視角) ---
                # A. 先將原始數據轉換為 dB 浮點數
                df[y_col] = df[y_col] / 1.0
                
                # B. 數值反轉 (將負的功率趨勢轉為正的損耗趨勢)
                # 這裡我們使用一個補償係數 (Offset)
                # 根據你的圖 2，起始點約在 14dB，而目前圖 1 起始點在 -51dB 左右
                # 所以補償值約為 14 - (-51) = 65
                
                offset_value = 65.0 
                df[y_col] = df[y_col] + offset_value

                print(f"✅ 數值校正完成。起始點現在約為: {df[y_col].iloc[0]:.2f} dB")
                
                # --- 5. 視覺化繪圖 ---
                plt.figure(figsize=(12, 6))
                plt.plot(df[x_col], df[y_col], color='blue', label='Corrected Trace', linewidth=1)
                
                plt.title(f"OTDR Trace: {os.path.basename(target_file)}")
                plt.xlabel(f"{x_col} (km)")
                plt.ylabel(f"{y_col} (dB)")
                plt.grid(True, linestyle='--', alpha=0.6)
                plt.legend()
                plt.show()
                # --- 6. 儲存為 JSON (供 AI 模型使用) ---
                json_path = target_file.replace('.sor', '.json')
                df.to_json(json_path, orient='records')
                print(f"✅ 已儲存校正後的數據至: {json_path}")
                
            else:
                print("❌ DataPts 區塊內沒有點位數據。")
        else:
            print("❌ 在檔案中找不到 DataPts 區塊。")
            print("現有的區塊名稱有:", [b.get('name') for b in sor_data])

    except Exception as e:
        print(f"❌ 執行過程發生錯誤: {e}")

# 執行主程式
process_otdr()
```
